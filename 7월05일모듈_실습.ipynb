```python
#가변 인자를 받는, 함수
#def 함수이름(파라미터<전달값>, * ):    #*<< 가변인자 표시

def profile(name,age, *lnagug): #langug에 여러 전달 값을 받을 수 있음.
  print(f"이름 : {name} \n나이 : {age}\n언어 :",end= " ")
  for i in lnagug:
    print(i, end=" ")

profile("응애","dmddo","A","B","C","D")
```

```python
#입력으로 들어오는 모든 수의 평균을 계산하는 함수

def avrgs(*number):
  sums = 0
  for i in number:
    print(i,"들어옴", end = " ")
    sums += i
  return sums/len(number)

print("\n\n\n",avrgs(2,2))
```

```python
# 반복문으로 n!반환 함수

def factorial_iteration(n):
  sum = 1
  for i in range(1,n+1):
    sum *= i
  return sum
def factorial_iteration_a(n):
  sum = 1
  while(n>0):
     sum *= n
     n-=1
  return sum

print(factorial_iteration(10))

print(factorial_iteration_a(10))
```

```python
def recursive_factorial(i):
  if i == 100:
    return
  else:
    print(f"{i}번째 재귀함수에서 {i+1} 함수를 호출 합니다.")
    return recursive_factorial(i+1)
  print(f"{i}번째 재귀함수를 종료합니다.")

recursive_factorial(1)

print("\n\n===================================\n\n")

def recursive_factorial(i):
  if i == 1:
    return
  print(f"{i}번째 재귀함수에서 {i-1} 함수를 호출 합니다.")
  return recursive_factorial(i-1)

  print(f"{i}번째 재귀함수를 종료합니다.")
recursive_factorial(100)

print("\n\n===================================\n\n")

def recursive_factorial(n):
  sum =0
  if n == 1:
    return 1
  else:
    sum = recursive_factorial(n-1)*n
  return sum

print(recursive_factorial(5))

```

```python
# 최대 공약수 계산 프로그램

def gcd(x, y):
  i = min(x,y)
  while(1):
    if x % i == 0 and y % i == 0:
      return i
    else:
      i = i -1
x = int(input("x 입력 :"))
y = int(input("y 입력 :"))
print(gcd(x,y))
```
