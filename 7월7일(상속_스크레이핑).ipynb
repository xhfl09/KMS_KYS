```python
#Super 사용 코드
class MyScore:
    def __init__(self, kor, eng):
        self.kor = kor
        self.eng = eng

    def getTit(self):
        return (self.kor + self.eng)

class MyScore_sub(MyScore):
    def __init__(self, kor, eng, math):
        self.math = math
        super().__init__(kor, eng)

    def getTit(self):
        return (self.kor + self.eng + self.math)

if __name__ == '__main__':
    print(MyScore(57,67).getTit())
    print(MyScore_sub(57,67,77).getTit())

print('\n===========================================\n')

class MyScore:
    def __init__(self, kor, eng):
        self.kor = kor
        self.eng = eng

    def getTit(self):
        return (self.kor + self.eng)

class MyScore_sub(MyScore):
    def __init__(self, kor, eng, math):
        MyScore.__init__(self, kor, eng)
        self.math = math

    def getTit(self):
        return (self.kor + self.eng + self.math)

if __name__ == '__main__':
    print(MyScore(57,67).getTit())
    print(MyScore_sub(57,67,77).getTit())
```

#### 피보나치 수열

```python
# 재귀함수 호출 일반
def fibo_self(n):
  if n == 1 or n==2:
    return 1
  else:
    return fibo_self(n-1)+fibo_self(n-2)

print(fibo_self(10))

#메모이제이션으로 제귀함수 효율화
def fibo_memo(n):
  d = [0]*n*10 #메모 공간 확보
  d[1] = 1
  d[2] = 1

  for i in range(3,n+1):
    d[i] = d[i-1] +d[i-2]
  return d[n]

print(fibo_memo(1000))
```

# 스크레이핑

```python
import requests

res=requests.get("https://media.naver.com/press/032")
print(res.status_code)
len(res.text)

from bs4 import BeautifulSoup

soup = BeautifulSoup(res.text,'lxml') # lxml << html 해체 특화?
print(soup.title,"\n\n") #html 가져옴
print(soup.title.get_text(),"\n\n=======================\n\n") #태그 안에 있는 내용을 가져옴
```

```python

#div 내용
print(soup.div,"\n\n")
print(soup.div.attrs,"\n\n") # 속성 가져오기
print(soup.div.get_text(),"\n\n======================\n\n") #내용
```

```python

print(soup.a,"\n\n")
print(soup.a.attrs,"\n\n")
print(soup.a["href"],"\n\n")
print(soup.a.get_text(),"\n\n======================\n\n")
```

```python
print(soup.find('div')) #찾기
```

```python
div = soup.find("div",attrs={"class":"press_container"}) #soup에 있는 press_container class 찾음
rank1 =div.find('a') # div 안에 있는 a (a태그)를 find
rank1.next_sibling # a 와 같은 위치에 있는 값을 찾음
div.find_all('a') #전체 a 태그를 찾음
```

```python
# tag와 attr 정보로 주요기사 title 가져오기

lst_li = soup.find_all('li', attrs = {'class' :'ofra_list_item as_thumb'})
print(len(lst_li))

for i,li in enumerate(lst_li):
  print(f"{i+1}번째 뉴스 ",li.find('div', attrs ={'class':'ofra_list_tx_headline'}).get_text())
  print(li.find('a', attrs ={'class':'ofra_list_item_link _es_pc_link'})["href"],'\n')
```

```python
main_news = soup.find_all('li', attrs = {'class' :'press_news_item'})
print(len(main_news))

for i , li in enumerate(main_news):
   print(f"{i+1}번째 뉴스 ",li.find('strong').get_text())
   print(li.find('a', attrs ={'class':'press_news_link _es_pc_link'})["href"],'\n')
```
