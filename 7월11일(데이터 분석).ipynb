```python
dset={"영미":{"국어":80, "수학":90, "음악":100},
      "두현":{"국어":100, "수학":98, "음악":90},
      "준이":{"국어":95, "수학":93, "음악":92}}
```

```python
score = pd.DataFrame(dset).T
display(score)
```

```python
score["영어"] = [90,95,92,98]
score.loc['은연'] = [80,100,80,98]
# 각 학생의 평균 점수 구하기
score["평균"] = (score["국어"] + score["수학"] + score["음악"] + score["영어"]) / 4

# 각 과목의 평균점수 구하기
score.loc['과목평균'] = score.mean(axis = 0)

# 영미, 두현, 준이 각 과목에서 받은 점수와 과목 평균과 편차
score.loc['영미 편차'] = score.loc['영미'] - score.loc['과목평균']
score.loc['두현 편차'] = score.loc['두현'] - score.loc['과목평균']
score.loc['준이 편차'] = score.loc['준이'] - score.loc['과목평균']
score.loc['은연 편차'] = score.loc['은연'] - score.loc['과목평균']
display(score)
```

# sklearn

```python
from sklearn import datasets

data = datasets.load_iris()
data
```

```python
#1.  data 추출 2차원
iris_data = data['data']
print(iris_data)

#2. target 추출
target_data = data['target']
print(target_data)

#3. fucher name 추출
feature_names= data['feature_names']
print(feature_names)

# target_names 1차원
target_names = data['target_names']
print(target_names)
```

```python
# 데이터 프레임 만들기
df = pd.DataFrame(data.data, columns=data.feature_names)
display(df)

```

```python
# iris 데이터의 평균 구하기
a = df.mean()
display(a)
print('\n\n\n==================================================\n')

# iris 데이터에 target 정보 추가하기
df['target'] = data.target
display(df)
print('\n\n\n==================================================\n')

# iris target의 0-> 'setosa', 1-> 'versicolor', 2-> 'virginica'로 변환
df['target'] = df['target'].map({0:"setosa", 1:"versicolor", 2:"virginica"})
display(df)
print('\n\n\n==================================================\n')

#각각의 평균 구하기
setosa=df[df['target'] == 'setosa'].mean()
versicolor= df[df['target'] == 'versicolor'].mean()
virginica = df[df['target'] == 'virginica'].mean()
print('\n\nsetosa: ',setosa,'\n\n\nversicolor: ',versicolor,'\n\n\nvirginica :',virginica)
```
