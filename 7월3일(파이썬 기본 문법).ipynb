### 반복문 (for)
  -

```python
fruites = ["apple","watermelon", "bannana", "perar"]
for fruite in fruites:
  print("이거는",fruite,"입니다.")

for i,a in enumerate(fruites):
  print(i," 번째",a)
```

```python
print("range(10)")
for i in range(10):
  print(i)
print("\n===================================\n")
print("range(5,10)")
for i in range(5,10):
  print(i)
print("\n===================================\n")
print("range(0,10,2)")
for i in range(0,10,2):
  print(i)
```

```python
# 구구단 프로그램
for i in range(2,10):
  print("\n===================================\n")
  for j in range(2,10):
    print(i,"X",j,"=",i*j)
```

```python

point = [10,20,30,40,50,42,31,54,32,12]
standard = 50
for number,poin in enumerate(point):
    if poin>=standard:
        print(number,"합격입니다.\n")
    else:
        print(number,"불합격입니다.\n")
```

```python
n = 1260
count = 0

coin_types = [500,100,50,10]

for i in coin_types:
  count += n//i
  n %= i


print(count,"의 동전을 사용했습니다")
```

```python
# 500, 300, 100, 50, 10 화폐가 있을때
#가장 적은 수의 화폐로 돈을 거슬러주는 프로그램을 작성하시오

# 거스름돈 입력
n = int(input("거슬러 줘야 하는 돈 : "))
# 화폐 단위
array = [500, 300, 100, 50, 10]
# 거스름돈 개수
count = 0

for coin in array:
    count += n // coin
    n %= coin

print(count)
```

```python
# 2진법을 10진법으로
coises = int(input("어떤 진법을 입력 하실 건가요?(2,8,16,32)"))
n = int(input("진법을 입력 하시오 : "))
i=0
result=0
while n>0:
  result += (n % 10) * (coises ** i)
  n //= 10
  i+=1
print(result)
```

### 파이썬의 특징
1. 파이썬 문법은 간단, 간결하고 보기 좋음. (들여쓰기 구조)
2. 파이썬의 자료형은 동적이다. (변수 앞에 생성자X)
3. 파이썬의 문자열은 유니코드다
4. 함수, 객체가 모여 모듈 || 모듈이 모여 파이썬 패키지이다.
 - 모듈 : 파일 단위로 구성된 별도의 네임스페이스



### 파이썬의 변수 연산자
1. 파이썬의 식별자 선언 규칙 (쿨래스 명을 제외하고는 대부분 소문자)
2. 변수 (=변하는 수)
 - 숫자,문자,클래스 객체를 나타냄
 - 저장되는 값은 변경 가능(이름은 유지)
 - 데이터의 형식이 정해질 떄 마다 메모리가 정해짐
3. 파이썬의 자료형
 - (위에 사진)
 - type() 함수를 사용하여 확인 가능
 - 비교 연산자 =, !=, <, <=, >=, >, is, is not, in, not in
 - 연산자 (위 사진)
 - 조건문 if, elif, else
