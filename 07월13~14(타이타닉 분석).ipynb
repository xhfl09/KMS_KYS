```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# /content/titanic_train.csv파일을 데이터프레임 객체로 불러오기
# head() 확인
# 데이터 수:, 특성수:, 각 특성의 데이터 타입, 결측치확인: , 중복확인:
```

```python
#타이타닉 불러오기
#미싱데이터 없애기 1. 결측치가 500개 이사인 열 없에기, 2. 결측치가 하나라도 있는 행 모두 앖애기
#중복데이터 제거
titanic=pd.read_csv("/content/titanic_train.csv")
titanic.info()
```

```python
titanic.drop("Cabin", axis=1, inplace=True)
titanic.dropna(how="any", inplace=True)
titanic.drop_duplicates( inplace=True)
```

```python
titanic.info()
```

```python
titanic["Pclass"].value_counts()
```

```python
titanic.head()
```

```python
titanic=titanic[["Survived",	"Pclass",	"Sex",	"Age" ,"Fare"]]
titanic.groupby([ "Survived", "Pclass"]).size()
```

```python
titanic.groupby([ "Survived", "Pclass"]).size().unstack()
```

```python
titanic.pivot_table("Age",index=["Survived"], columns=["Pclass"], aggfunc="count", margins=True)
```

```python
titanic["Survived"].groupby(titanic["Pclass"]).mean()
```

```python
titanic["Age"].value_counts().sort_index()
```

```python
names = [10,20,30,40,50,60,70,80]
titanic["age_interval"] = pd.cut(titanic["Age"],8, labels = names)
titanic.info()
```

```python
titanic["age_interval"].value_counts()
```

```python
titanic.pivot_table("Age",index=["Survived"], columns=["age_interval"], aggfunc="count")
```

```python
titanic["Survived"].groupby([titanic["Sex"], titanic["age_interval"]]).mean().unstack()
```
