```python
import requests

res=requests.get("https://search.daum.net/search?w=tot&q=2022%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR")
print(res.status_code)
soup=BeautifulSoup(res.text, "lxml")
images=soup.find_all("img", attrs={"class":"thumb_img"})
len(images)
import
list_li = soup.find('ol', attrs = {'class':'type_plural list_exact movie_list'})

li_list=list_li.find_all("li")
print(len(li_list))

#print(img_res.status_code)


for i , li in enumerate(li_list):
  print(f'{i+1}번째 영화', li.find('a', attrs={'class':'tit_main'}).get_text())
  print('링크 : ',"https://search.daum.net/search"+li.find('a')["href"],'\n=================================\n')
  img_url = li_list[i].find('img',attrs={"class":"thumb_img"})['src']
  img_res = requests.get(img_url)
  with open(f'응애{i+1}.jpg','wb')as f:
    f.write(img_res.content)
  if i ==5:
    break

#print(f"{i+1}번째 뉴스 ",li.find('div', attrs ={'class':'ofra_list_tx_headline'}).get_text())
```
