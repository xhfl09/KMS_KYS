```python
import numpy as np
dset={"영미":{"국어":80, "수학":90, "음악":100},
      "두현":{"국어":100, "수학":98, "음악":90},
      "준이":{"국어":95, "수학":93, "음악":92}}
weight=[1, 1.2, 2.0]
# 학생의 총점(가중치 고려)

# 학생별 총점
# 학생별 순위(토탈기준)
for name, points in dset.items():
    total=0
    for i, point in enumerate(points.values()):
        total+= point*weight[i]
    print(name, total)
############################################################# 강사
dset={"영미":{"국어":80, "수학":90, "음악":100},
      "두현":{"국어":100, "수학":98, "음악":90},
      "준이":{"국어":95, "수학":93, "음악":92}}
weight=[1,1.2, 2.0]
#학생의 총점(가중치 고려)
score=[]
for i in dset.values():
  score.append(list(i.values()))
score_arr = np.array(score)
weight_arr = np.array(weight)

print(score_arr*weight_arr)

total_sum = (score_arr*weight_arr).sum(1)
print(total_sum)
print(total_sum.max())    # 가장 큰 값을 찾음
print(total_sum.argmax()) # 가장 큰 값이 위치한 인덱스를 찾음
id = total_sum.argmax()
print(list(dset.keys())[id],':',total_sum.max())
```

```python
# 계단을 올라간다
# randint(0,2)
# 0 이 나오면 올라가지 않음
# 1 이 나오면 올라감
# 100 번 반복
arr = np.random.randint(0,2,100)
print(arr)
max = -999
cnt = 0
for i in arr:
    if i == 1:
      cnt += 1
    else:
      cnt-=1
    if max<cnt:
      max = cnt

print('최대로 높이 올라갔을 때',max)
print(cnt)
```

### Seires : 벡터 형태의 자료구조

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

obj = pd.Series([4, 7, -5, 3])
print(obj) # 좌측 인덱스, 우측 값
print(type(obj))

obj = pd.Series([4,7,-5,3], index = list('abcd')) # 인덱스 변경
print(obj)
print(obj.values)
print(obj.index)
```

```python
print(obj['a'],'\n') # 하나 << 숫자만
print(obj[['b','a']],'\n') # 다수 << 시리즈 자체를 가져옴
print(obj['b':'d'],'\n') # d가 포함이 되어 가져옴
print(obj>=4,'\n')
```
