### 슬라이싱


```python
from ctypes import sizeof
a = [1,2,3,4,5]
print(a[0:3]) # 0,1,2 번 인덱스 까지 출력

a.append(['a','b','c'])
print(a)

print(a[-1]) # 마지막 원소 출력, 이 때 마지막 원소는 리스트 임으로 리스트가 출력

print(a[-1][1:])

print(a[:]) # 전체 출력

print(a[::2]) # 스탭 추가

print(a[::-1]) #역순 출력

a.insert(1, 100) # 인덱스 1 자리에 100을 넣어라, (교체 아님, 중간에 추가하는 것임, 고로 인덱스의 길이가 늘어남에 주의)
print(a)

i = [2,3,4,5,6,7,8] #i을 이용하여 i의 i의 원소에 제곱을 한 리스트를 생성
temp = []
for j in i:
  temp.append(j**2)
print(temp)

#liste comprehensions 방식을 이용
i3 = [num**2 for num in i]
print(i3)
```

#### liste comprehensions

```python
b = [y for y in range(2000, 2100)] #liste comprehensions
print(b)



#liste comprehensions 사용하지 않는 코드
b = []
for y in range(200,2100):
  if y%4==0:
    b.append(y)
print(b)

#liste comprehensions 사용한 코드
b = [y for y in range(2000,2100) if y%4==0]
print(b)



#liste comprehensions 사용하지 않는 코드
result = []
for i in range(5):
  for j in range(3):
    result.append((i,j))
print(result)

#liste comprehensions 사용한 코드
result = [(i,j) for i in range(5) for j in range(3)]
print(result)
```

```python
jin= int(input("진법을 입력 하시오 (2,8,16,32) :"))
numb = int(input("입력 :"))


def dsa(jin,numb):
  out_numb = numb
  i = 0
  out_num =0
  while numb>0:
    out_num += (numb%10)*(jin**i)
    numb //=10
    i+=1
  return out_num

print(jin,"진법",numb,"의 10진수 변환은",dsa(jin,numb),"입니다.")
```

```python
# 두 점 사이의 거리를 구하는 함수를 작성
# 두 점은 사용자가 입력

import math

def distance(x1,y1,z1,x2,y2,z2):
    return math.sqrt((x2-x1)**2 + (y2-y1)**2 + (z1-z2)**2)

x1 = int(input("1번 x점을 구하시오 :"))
y1 = int(input("1번 y점을 구하시오 :"))
z1 = int(input("1번 z점을 구하시오 :"))
x2 = int(input("2번 x점을 구하시오 :"))
y2 = int(input("2번 y점을 구하시오 :"))
z2 = int(input("2번 z점을 구하시오 :"))
print(distance(x1,y1,z1,x2,y2,z2))

x = [1,2,3]
y = [1,2]
def dsis(x,y):
  if len(x) != len(y):
    print("계산 불가")
    return 0
  result = 0
  for i in range(len(x)):
    result += (x[i]-y[i])**2
  return result
print(dsis(x,y))
```

```python
#표준 체중을 구하는 프로그램
# 남 : 키(m) * 키(m) * 22
# 여 : 키(m) * 키(m) * 21
# 조건 1 : 함수명은 std_weight
# 조건 2 : 사용자 입력 값은 키(cm), 성별(gender)
# 조건 3 : 표준 체중은 소수점 둘째자리까지 표시

# (출력 예제)
# 키 175cm 남자의 표준 체중은 67.38kg 입니다.
import math

def avares(k , s):
  k /= 100
  if s == '남자':
    return round(k * k * 22, 2)
  elif s == '여자':
    return round(k * k * 21,2)

ks = int(input("키 입력 :"))
se = str(input("성별 입력 : (남자 / 여자)"))

print(f"키 {ks}의 {se}의 표준 체형은 {avares(ks,se)} 입니다.")
```

```python
# n! 프로그램


def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

num = int(input("숫자 입력 : "))
print(factorial(num))
```
