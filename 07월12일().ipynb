```python
# target 정보 뺴오기
display(iris[iris['target'] == 0])
# target ==0 인 데이타의 평균
print('\n\n',iris[iris['target']==0].mean(0),'\n')
iris0_mean = iris[iris['target']==0].mean(0).drop('target')
print(iris0_mean,'\n')
```

```python
# target 정보 뺴오기
# target ==1 인 데이타의 평균
print('\n\n',iris[iris['target']==1].mean(0),'\n')
iris1_mean = iris[iris['target']==1].mean(0).drop('target')
print(iris1_mean,'\n')
```

```python
# target 정보 뺴오기
# target == 2 인 데이타의 평균
print('\n\n',iris[iris['target']==2].mean(0),'\n')
iris2_mean = iris[iris['target']==2].mean(0).drop('target')
print(iris2_mean,'\n')
```

```python
print(iris_mean,'\n')
print(iris0_mean,'\n')
print(iris1_mean,'\n')
print(iris2_mean,'\n')
```

```python
df=pd.DataFrame()
df["iris_mean"]=iris_mean
df["iris0_mean"] = iris0_mean
df["iris1_mean"] = iris1_mean
df["iris2_mean"] = iris2_mean
display(df)
```

```python
# iris0_mean,iris1_mean,iris2_mean 의 평균이 iris_mean 동일한지
import numpy as np
print(np.round(df.iris_mean,2)==np.round((df.iris0_mean+	df.iris1_mean +	df.iris2_mean)/3,2),'\n')
# 넘파이로 사용하는 이유(소수점이 있을경우) 컴퓨터는 2진수 계산을 함으로 2진수 계산시 달라질 수 있음 대표적으로
# 1.0+1.2 == 1.2 fales 경우임, 고로 소수점 자리수를 정해주는 걸 추천함.

print(np.round(df.iris_mean)==np.round(df.drop("iris_mean", axis=1,inplace=False).mean(1)))

```

```python
np.round(df.iris_mean,2)==np.round(df[["iris0_mean",	"iris1_mean",	"iris2_mean"]].mean(1),2)
```

```python
# iris0_mean,iris1_mean,iris2_mean
df.sub(df.iris_mean,axis=0)
```

```python
#7. 1호선, 2호선, 3호선, 4호선의 이름을 line1,line2, line3, line4로 변경
display(subway)
#subway['노선명'] = subway['노선명'].map({"1호선":"line1", "2호선":"line2", "3호선":"line3", "4호선":"line4"})
subway['노선명'].replace(['1호선','2호선','3호선','4호선'],['line1','line2','line3','line4'])
subway['노선명'].unique()


subway.loc[subway['노선명']== '1호선', '노선명']='line1'
subway.loc[subway['노선명']== '2호선', '노선명']='line2'
subway.loc[subway['노선명']== '3호선', '노선명']='line3'
subway.loc[subway['노선명']== '4호선', '노선명']='line4'

#display(subway)
```

```python
#8. 날짜 20210201형태를 2021-02-01로 변경=< 함수 정의후 Series.apply() 사용
def date_format(d):
  d = str(d)
  return str(d[:4]+ '-' + d[4:6] + '-' +d[6:])

subway['사용일자'] = subway['사용일자'].apply(date_format)
subway['등록일자'] = subway['등록일자'].apply(date_format)

display(subway)
```

```python
subway = subway.set_index('사용일자')
display(subway.head())
```

```python
print(subway['승차총승객수'].max())
print(subway['승차총승객수'].min())
print(subway['승차총승객수'].argmax())
print(subway['승차총승객수'].idxmax())
```

```python
#등록일자 삭제
subway = subway.drop('등록일자', axis = 1)
display(subway)
```

## 데이타 합치기


### merge

```python
df1=pd.DataFrame({"이름":list("ABC"), "성별":["남자","여자","남자"]})
df1
```

```python
df2=pd.DataFrame({"이름":list("ABD"), "키":["180","150","170"]})
df2
```

```python
# 이름이 같은 컬럼을 기준으로 머지함
pd.merge(df1,df2) #default는 inner join
```

```python
pd.merge(df1,df2, how="left")
```

```python
pd.merge(df1,df2, how="right")
```

```python
df3=pd.DataFrame({"이름":list("ABC"), "성별":["남자","여자","남자"],
                  "학번":[1,2,3]})
df4=pd.DataFrame({"이름":list("김박최"), "키":["180","150","170"],
                  "학번":[1,2,4]})

print(df3,'\n\n',df4)
```

```python
pd.merge(df3,df4)
```

```python
#pd.merge(df3,df4,how = "outer") # << new데이타를 만듬
pd.merge(df3,df4, on = '학번',how = "outer") # 중심설정 on = ~

```

```python
score1=pd.DataFrame({"이름":list("DACBE"), "영어":[90, 80, 70, 65, 100],
              "수학":[100,90,90,65,70],"학번":[5,7,8,4,2]} )
score2=pd.DataFrame({"name":list("DACBE"), "국어":[90, 80, 70, 65, 100],
              "id":[5,7,8,4,2]})
display(score1)
display(score2)
```

```python
# score1, score2 두개의 데이타를 합치자
pd.merge(score1, score2,left_on="학번", right_on="id")
```

### concat
- 축따라 이어 붙이기, 여러개 이어붙이기 가능 (입력은 리스트)

```python
df1 = pd.DataFrame({'이름':['A','B','C','D','E'], '키':[180,160,150,170,170], '나이':[23,46,33,65,28], '주소':["서울","부산","대구","광주","제주"]})
df2 = pd.DataFrame({'이름':['F','G','H','I','J'], '키':[190,180,150,160,170], '나이':[22,18,87,45,74], '주소':["전주","서울","수원","부산","인천"]})

display(df1)
print('------------------------------')
display(df2)
```

```python
pd.concat([df1,df2])
```

```python
display(pd.concat([df1,df2],axis=0))

display(pd.concat([df1,df2],axis=1))
```

```python
df3 = pd.concat([df1,df2],axis=0,ignore_index=True)
display(df3)
```
